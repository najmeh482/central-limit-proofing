```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Generate 1000 values from a Poisson distribution with lambda=5
poisson_data = np.random.poisson(lam=5, size=1000)

# Initialize a list to store the means
means = []

# Repeat the process 1000 times
for _ in range(1000):
    # Choose 100 random numbers from the poisson_data
    sample = np.random.choice(poisson_data, size=100, replace=False)
    # Calculate the average of the sample
    sample_mean = np.mean(sample)
    # Store the average
    means.append(sample_mean)

# Plot the histogram of the means
plt.hist(means, bins=100, alpha=0.7, color='green')
plt.title('Histogram of Sample Means from Poisson Distribution')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()
```
